```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns ##Country code is same column in both the tables can we combine these two?
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
##for datatypes
final_df.dtypes
```

```python
final_df.Country.value_counts() ## give the count of countries
```

```python
country_names=final_df.Country.value_counts().index ##To store the country names
country_numbers=final_df.Country.value_counts().values
```

```python
##To see the numbers of the top 3 countries
plt.pie(country_numbers[:3], labels=country_names[:3], autopct='%1.2f%%')
```

# **Observations:**
1.   Zomato Maximum records or transaction are from India
2.   after that USA and UK


```python
final_df.columns
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# **Observations**


1.   When Rating is between 4.5 to 4.9 --> Excellent
2.   When Rating is between 4.0 to 4.4 --> Very Good
3.   When Rating is between 3.5 to 3.9 --> Good



```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue = "Rating color", data=ratings, palette=['lavender', 'red', 'orange', 'yellow', 'green', 'green'])
```

# **Observations**
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
##Count plot
sns.countplot(x='Rating color', data=ratings, palette=['lavender', 'red', 'orange', 'yellow', 'green', 'green'])
```
