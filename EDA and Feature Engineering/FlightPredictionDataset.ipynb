# **Flight Prediction (EDA + Feature Engineering)**

```python
#Importing basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  # we dont need to write plot.show it is shown automatically
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
df = pd.concat([train_df, test_df], ignore_index=True)
df.head()
```

```python
df.info()
```

```python
## Creating derived Feature from DOJ
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df.drop('Date_of_Journey', axis=1, inplace=True)
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.head()
```

```python
df.info()
```

```python
# df['Arrival_Time'].str.split(' ').str[0]
# Using Lambda
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
df.head()
```

```python
df.isnull().sum()
```

```python
##converting to int the arrival time
df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x: x.split(':')[0])
df['Arrival_Minute'] = df['Arrival_Time'].apply(lambda x: x.split(':')[1])
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
df.drop('Arrival_Time', axis=1, inplace=True)
df.head(1)
```

```python
##Doing same thing for dept time
df['Dept_Hour'] = df['Dep_Time'].apply(lambda x: x.split(':')[0])
df['Dept_Minute'] = df['Dep_Time'].apply(lambda x: x.split(':')[1])
df['Dept_Hour'] = df['Dept_Hour'].astype(int)
df['Dept_Minute'] = df['Dept_Minute'].astype(int)
df.drop('Dep_Time', axis=1, inplace=True)
df.head(1)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4})

df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route', axis=1, inplace=True)
df.head()
```

```python
#Handling duration
df['duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df[df['duration_hour']=='5m']
```

```python
df.drop(6474, axis=0, inplace=True)
df.drop(13343, axis=0, inplace=True)
```

```python
df['duration_hour']=df['duration_hour'].astype('int')
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
df.head(1)
```
