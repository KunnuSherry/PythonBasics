# **To Delete rows and columns to handle null values bold text**

```python
#Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
df = pd.read_csv('train.csv')
```

```python
df.shape #for rows and cols
```

```python
#Showing all rows and columns
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df.head(6)
```

```python
df.info()
```

```python
#Column wise null values
df.isnull().sum()
```

```python
#Visulalising the null values
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
#Null Percentage
null_per = df.isnull().sum()/df.shape[0]*100
null_per
```

```python
#Keys of the columns whose nnull percentage is greater than 17 percent
drop_cols = null_per[null_per>17].keys()
drop_cols
```

```python
df_dropped_cols = df.drop(columns=drop_cols)
df_dropped_cols.shape
```

```python
#Now we can see the null values have significantly decreased
sns.heatmap(df_dropped_cols.isnull())
```

```python
#Now removing the rows with null values
df_dropped_rows = df_dropped_cols.dropna()
df_dropped_rows.shape
```

```python
#Let's see the heatmap
sns.heatmap(df_dropped_rows.isnull())
```

```python
#None of the values is now null
df_dropped_rows.isnull().sum().sum()
```

```python
#Visualising the changes there are not much changes
import matplotlib.pyplot as plt
import seaborn as sns

num_var = df_dropped_rows.select_dtypes(include=['int64', 'float64']).columns

rows = (len(num_var) // 4) + 1  # Dynamic rows for subplots
plt.figure(figsize=(20, 5 * rows))

for i, var in enumerate(num_var):
    plt.subplot(rows, 4, i+1)
    sns.histplot(df[var], bins=20, kde=True, color='blue', alpha=0.5, label='Original')
    sns.histplot(df_dropped_rows[var], bins=20, kde=True, color='red', alpha=0.5, label='After Drop')
    plt.title(var)
    plt.legend()

plt.tight_layout()
plt.show()
```
