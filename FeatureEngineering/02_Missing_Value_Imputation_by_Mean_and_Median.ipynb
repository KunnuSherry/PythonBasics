# **Missing Value Imputation by Mean and Median**

```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
dataset_path = r"https://drive.google.com/uc?export=download&id=1BiGZSedP4BIIuTbVTBodOhVgFImaz08c"
df = pd.read_csv(dataset_path)
df.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
missing_value_per = df.isnull().sum()/df.shape[0] * 100
missing_value_per
```

```python
missing_value_clm_gre_20 = missing_value_per[missing_value_per > 20].keys()
missing_value_clm_gre_20
```

```python
#Dropped columns with null values greater than 20 percent
df2_drop_clm = df.drop(columns=missing_value_clm_gre_20)
df2_drop_clm.shape
```

```python
#handling int and float
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
#Visualising through the heatmap the missing values
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
#Getting cols with null vals
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

# **There is a thing how we should fill with mean or median:**
1. If we have gaussian dist. like in 1st we can use either of mean or median.
2. For non-gaussian it's better to use median.

```python
#Plotting the dist. for all three cols
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df3_num[var], hist=False, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
#replacing the null values with mean
df4_num_mean = df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
#Replacing the null values with median
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
#Plotting the changes
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1);
    sns.distplot(df3_num[var],hist=False, kde_kws={'linewidth': 2, 'color':'red'}, label='Original Data')
    sns.distplot(df4_num_mean[var],hist=False, kde_kws={'linewidth': 2, 'color':'green'}, label='Mean Data')
    sns.distplot(df5_num_median[var],hist=False, kde_kws={'linewidth': 2, 'color':'black'}, label='Median Data')
```

```python
#Seeing outliers
for i, var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
#Observing changes
df_concat = pd.concat([df4_num_mean, df5_num_median], axis=1)
df_concat[df_concat.isnull().any(axis=1)]
```
