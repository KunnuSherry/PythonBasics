```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
#Loading dataset
train = pd.read_csv('train.csv')
train.head()
```

```python
#selecting numerical variables
nums_vars = train.select_dtypes(include=['int64','float64']).columns
nums_vars
```

```python
train[nums_vars].isnull().sum()
```

```python
imputer_mean = SimpleImputer(strategy='mean')
# imputer_mean = SimpleImputer(strategy='constant', fill_value=0)
imputer_mean.fit(train[nums_vars])
imputer_mean.statistics_
```

```python
train[nums_vars]=imputer_mean.transform(train[nums_vars])
```

```python
train[nums_vars].isnull().sum().sum()
```

# **For categorical Variables**

```python
cat_vars=train.select_dtypes(include=["O"]).columns
cat_vars
```

```python
train[cat_vars].isnull().sum()
```

```python
imputer_mode = SimpleImputer(strategy='most_frequent')
#imputer_mean = SimpleImputer(strategy='constant', fill_value=99)
imputer_mode
```

```python
imputer_mode.fit(train[cat_vars])
```

```python
train[cat_vars] = imputer_mode.transform(train[cat_vars])
```

```python
train[cat_vars].isnull().sum().sum()
```
