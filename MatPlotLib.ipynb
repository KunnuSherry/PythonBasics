```python
import matplotlib.pyplot as plt                                        
import numpy as np
%matplotlib inline                                                      
x = np.arange(1,10)
y = np.arange(11,20)

plt.scatter(x,y,c='r')                           #Plotting the graph
plt.xlabel('X-Axis')                             
plt.ylabel('Y-Axis')
plt.title('MatPlotLib')
# plt.show()                                     #we need to write this if % line is not there
```

```python
y = x*x
plt.plot(x, y, 'ro--', linewidth = 2, markersize=12)      #we can also use linestyle
```

```python
#subplot means showing multiple graphs at one canvas
plt.subplot(2,2,1)
plt.plot(x, x*x, 'g')
plt.subplot(2,2,2)
plt.plot(x, x*x*x, 'r--')
```

```python
#sin graph
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, "r--")
```

```python
#bar graphs and Histograms

import numpy as np
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
x = np.array([2,4,6,8])
y= np.array([12,14,16,18])
plt.bar(x,y,color="r")
x1 = np.array([2,3,4,5,2,3,3,4,1,100,23,21,33,45])
plt.subplot(1,2,2)
plt.hist(x1)
```

```python
#piecharts

import numpy as np
import matplotlib.pyplot as plt
labels = ['Python','Java','C','C++']
sizes = [215,130,245,210]
colors=['gold', 'yellow', 'blue', 'red']
explode = (0.1,0,0,0)        #distance from origin
plt.pie(sizes, explode=explode, labels = labels, colors=colors, shadow=True, autopct='%1.1f%%')
```
